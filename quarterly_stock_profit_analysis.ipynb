Data **visualization**

```python
df=pd.read_csv('total.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 0'])
```

```python
df.dtypes
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
plt.figure(figsize=(11, 8))
sns.lineplot(data=df, x="date", y="profit", hue="company", marker="o")

plt.title("Profit Over Time by Company")
plt.xlabel("Date")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df_ahpc = df[df['company'] == 'AHPC']
sns.barplot(
    data=df_ahpc,
    x="quarter",
    y="profit",
)

plt.title("Quarterly Profit for AHPC")
plt.xlabel("Quarter")
plt.ylabel("Profit")
plt.tight_layout()
plt.show()
```


**Shows spread, outliers, and volatility per company**

```python
plt.figure(figsize=(8, 6))
df_filtered = df[df['company'].isin(['AHPC', 'DHPL', 'BARUN'])]
sns.boxplot(
    data=df_filtered,
    x="company",
    y="profit",
)

plt.title("Profit Distribution by Company")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.tight_layout()
plt.show()
```

```python
top_companies = ["AHPC", "BHPC", "CHPC", "DHPC", "EHPC"]

# Filter dataframe
df_top5 = df[df["company"].isin(top_companies)]
total_profit = df.groupby("company")["profit"].sum()

# Filter out companies with non-positive total profit for the pie chart
positive_total_profit = total_profit[total_profit > 0]

plt.figure(figsize=(6, 6))
plt.pie(
    positive_total_profit,
    labels=positive_total_profit.index,
    autopct="%1.1f%%",
    explode=[0.1] * len(positive_total_profit),
    startangle=90,
    colors=sns.color_palette("Set2")
)
plt.title("Total Profit Share by Company (Positive Profits Only)")
plt.tight_layout()
plt.show()
```

```python
top_companies = ["AHPC", "SHPC", "MEN", "BARUN", "MHNL"]

# Filter dataframe
df_top5 = df[df["company"].isin(top_companies)]
total_profit_top5 = df_top5.groupby("company")["profit"].sum()

# Keep only positive profits
positive_profit_top5 = total_profit_top5[total_profit_top5 > 0]
plt.figure(figsize=(6, 6))
plt.pie(
    positive_profit_top5,
    labels=positive_profit_top5.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=sns.color_palette("Set2"),
    explode=[0.1 if v == positive_profit_top5.max() else 0 for v in positive_profit_top5]
)
plt.title("Total Profit Share of Top 5 Companies (Positive Profits Only)")
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(6, 5))
df.drop(columns=['Unnamed: 0'], inplace=True)
sns.heatmap(df.corr(numeric_only=True),annot=True , cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
